## Data Cleaning 

```python
# Check null values 
df.isnull().sum()
```

```python
# Fill null values in (Treatment Cost) 
df["Treatment_Cost"] = df["Treatment_Cost"].fillna(df["Treatment_Cost"].mean())
```

```python
# Fill null values in (Hospital Days)
df["Hospital_Days"] = df["Hospital_Days"].fillna(df["Hospital_Days"].mean())
```

```python
# Fill null values in (Insurance Provider)
df["Insurance_Provider"] = df["Insurance_Provider"].fillna("No Insurance")
```

```python
# Recheck null values after using fillna function 
df.isnull().sum()
```

```python
# To check for duplicates 
df["Patient_ID"].duplicated().sum()
```

```python
# Show duplicate records 
df[df["Patient_ID"].duplicated()]
```

```python
# Remove duplicate record 
df = df.drop_duplicates(subset = "Patient_ID")
```

```python
# Recheck for duplicate records 
df["Patient_ID"].duplicated().sum()
```

## Feature Engineering

```python
# Create a new column (Age Group)
df["Age_Group"] = [
    "Young" if age < 30 else
    "Adult" if age < 60 else
    "Senior" 
    for age in df["Age"]
]
```

```python
# To verify column is created 
df["Age_Group"].value_counts()
```

```python
# Create a new column (Cost Category)
df["Cost_Category"] = [
    "Low Cost" if cost < 50000 else
    "Medium Cost" if cost < 100000 else
    "High Cost"
    for cost in df["Treatment_Cost"]
]
```

```python
# To verify column is created 
df["Cost_Category"].value_counts()
```

```python
# Create a new column (Stay Category)
df["Stay_Category"] = [
    "Short Stay" if day <= 5 else
    "Medium Stay" if day <= 10 else 
    "Long Stay" 
    for day in df["Hospital_Days"]
]
```

```python
# To verify column is created 
df["Stay_Category"].value_counts()
```
